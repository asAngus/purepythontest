```python
import os
import pandas as pd
from prettypandas import PrettyPandas
```

```python
csvfile = '2012.csv'
df = pd.read_csv(csvfile, na_values=[''])
```

```python
df.head()
```

```python
dff = df.filter(items=[col for col in df.columns if col not in ['aa', 'ff']])
dff.head()
```

```python
pdff = PrettyPandas(dff)
```

```python
type(pdff.data)
```

```python
pdff.data.head()
```

```python
pdff
```

```python
pdff.total()
```

```python
df2 = pdff.data.filter(items=[col for col in df.columns if col not in ['Date']])
df3 = df2.head()
df3
```

```python
pdf3 = PrettyPandas(df3)
```

```python
pdf3
```

```python
# 按行相加，即 列值相加
pdf3.total(axis=0)
```

```python
# 按列相加
pdf3.total(axis=1)
```

```python
pdf3 = PrettyPandas(df3)
pdf3.total(title='total').average().median().max(title='你最大').min()
```
