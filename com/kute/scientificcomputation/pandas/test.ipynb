```python
# na_values：表明哪些值需要被视为 NAN
df = pd.read_csv('2012.csv', na_values=[''])
```

```python
# 过滤某些列
df1 = df.filter(items=[col for col in df.columns if col not in ['aa', 'ff']])
```

```python
df1
```

```python
# 选取多个列
df1[['Date', 'cc', 'dd']][:1]
```

```python
# 选取Date（注意值类型）值为  02/01/2012  和 01/11/2012
df1.query('Date == ["02/01/2012", "01/11/2012"]')
```

```python
df1[df1.cc.isin([0, 2])]
```

```python
# 过滤  cc 列出现次数 <= 1的数据
df1.groupby('cc').filter(lambda x: len(x) > 1)['cc'].value_counts()
```

```python
vc = df1['cc'].value_counts()
```

```python
# 过滤  cc 列出现次数 <= 1的数据
vc[vc > 1]
```
